2.   A último momento, vimos una cremona pequeña a \$ 900 y decidimos agregarla al pedido.
3.  En la panadería hay un descuento de \$150 ¿cómo podemos incorporarlo a nuestro programita?


No te olvides de colocar un comentario explicando qué haces en cada paso


*(Docena de facturas = (12u * $110))+ (Pan = (2u * $600))+(Cremona= (1u * $900)+(Descuento = (-150))= Total de la compra ($3270)*

```python
# Defino cada una de las variables involucradas en el proceso.
var1=('Factura')
var2=('Pan')
var3=('Cremona')
var4=('Descuento')
var5=('Total de la compra')
var6=110
var7=600
var8=900

# Calculo del total de la compra
importe=(var6*12)+(var7*2)+(var8*1)-(150)
# Castear: esto es para cambiar el tipo de variable str(var6*12),
# entonces tambien se resuelve como una funcion matematica primero resulto parentesis (var6*12),
# luego a este lo transforma en str (str(1320) = 1320), y luego (var1+'---'+1320)
# y al final llama a la funcion imprimir e imprime la suma de la concatenacion
# desp de todo este analisis y descripcion de lo que hice le agregue el signo $ y los .....
print(var1+' .............'+'$'+str(var6*12))
print(var2+'     .............'+'$'+str(var7*2))
print(var3+' ..............'+'$'+str(var8*1))
print(var4+' ...........'+'$'+str(-150))
print('--------------------------')
print(var5+' = '+'$'+str(importe))
```
